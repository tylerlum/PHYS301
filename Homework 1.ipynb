```python
import numpy as np
import matplotlib.pyplot as plt
np.seterr(divide='ignore', invalid='ignore')
```

```python
# Question 1 
# setup values
k = 8.99 * 10**(9)  # Nm^2/C^2
Q = 1 * 10**(-6)  # 1 uC
R = 0.1  # 10 cm
max_r = 0.5  # 50 cm
points_diff = 0.001  # dist between points in arrays

# set up arrays
r_i = np.arange(0, R, points_diff)  # inner radius
r_o = np.arange(R, max_r, points_diff)  # outer radius

V_i = (k * Q / R) * r_i / r_i  # inner voltage
V_o = k * Q / r_o  # outer voltage

# plot Voltage vs. Distance 
fig = plt.figure(figsize=(6,6))
plt.plot(r_i,V_i)
plt.plot(r_o,V_o)
plt.xlabel('Distance from Center of Sphere (m)')
plt.ylabel('Electric Potential (V)')
fig.savefig('Potential_From_Conducting_Sphere.png')
```

```python
# Question 2 
# setup integration function
from scipy.integrate import quad
def p(r):
    return 0.5*np.exp(-r**2)*r**2

# setup variables
R = 0.1  # 10 cm
max_r = 0.2  # 20 cm
points_diff = 0.001

# setup array and integration
all_r = np.arange(0, R, points_diff)
q_enclosed = []
for r in all_r:
    I = quad(p, 0, r)
    q_enclosed.append(I[0] * 4 * np.pi)

# plot Voltage vs. Distance 
fig = plt.figure(figsize=(6,6))
plt.plot(all_r, q_enclosed)
plt.xlabel('Distance from Center (m)')
plt.ylabel('Charge Enclosed (C)')

print("Total charge is", I[0] * 4 * np.pi, "Coulombs")
fig.savefig('Charge_Enclosed_In_exp_r_squared.png')
```
