```python
import numpy as np
import matplotlib.pyplot as plt

# Grid of x, y points
L=5
nx, ny = 100, 100
x = np.linspace(-L, L, nx)
y = np.linspace(-L, L, ny)
X, Y = np.meshgrid(x, y)

# demo of streamplot for a vector field
fig = plt.figure(figsize = (8,8))
plt.streamplot(x, y, np.sin(X), np.sin(Y), linewidth=1, density=2, arrowstyle='->', arrowsize=1.5);
```
