```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import rcParams
rcParams.update({'font.size':20}) #optional to change default font size

# define constants
k = 8.99 * 10**(9)  # N*m^2*C^-2
R = 1  # m
Q = 1 * 10**(-6)  # C
m = 0.1  # kg
x_0_1 = 1  # m
x_0_2 = 10  # m
v_0 = 0  # m/s
t_max = 1000  # s
dt = 1  # s

# set up arrays for position and time
time = np.arange(0, t_max, dt)
x_1 = []
x_2 = []
v_1 = []
v_2 = []

# define a function that returns the force or acceleration
def a(x):
    E = -k*Q*x/((x**2+R**2)**(1.5))
    return Q*E/m

# set initial position and velocity
x_1.append(x_0_1)
v_1.append(v_0)

x_2.append(x_0_2)
v_2.append(v_0)

# integrate equation of motion
for i in range(1, len(time)):
    # update position 
    next_x_1 = x_1[i-1] + dt*(v_1[i-1] + a(x_1[i-1])*dt/2)
    x_1.append(next_x_1)
    next_x_2 = x_2[i-1] + dt*(v_2[i-1] + a(x_2[i-1])*dt/2)
    x_2.append(next_x_2)
    
    # update speed
    next_v_1 = dt*(v_1[i-1] + a(x_1[i-1])*dt/2) + a(x_1[i])*dt/2
    v_1.append(next_v_1)
    next_v_2 = dt*(v_2[i-1] + a(x_2[i-1])*dt/2) + a(x_2[i])*dt/2
    v_2.append(next_v_2)

# plot trajectories
plt.figure(figsize=(6,6))
plt.plot(time, x_1)
plt.xlabel('Time (s)')
plt.ylabel('Position of Charge (x_0 = 1) (m)')

plt.figure(figsize=(6,6))
plt.plot(time, x_2)
plt.xlabel('Time (s)')
plt.ylabel('Position of Charge (x_0 = 10) (m)')
```
