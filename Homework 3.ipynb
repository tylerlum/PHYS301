```python
# Please run this cell twice to get plots. 
# Doesn't plot on first run for some reason. 
# If you know a reason why, please let me know :)
import numpy as np
import matplotlib.pyplot as plt

# Constants
k = 8.9 * 10**9  # Nm^2/C
a = 1  # m
b = 2  # m
q = 1  # C

# Grid of x, y points
L1 = 4
L2 = 10
nx, nz = 100, 100
x1 = np.linspace(-L1, L1, nx)
z1 = np.linspace(-L1, L1, nz)
X1, Z1 = np.meshgrid(x1, z1)

x2 = np.linspace(-L2, L2, nx)
z2 = np.linspace(-L2, L2, nz)
X2, Z2 = np.meshgrid(x2, z2)

# Electric field contributions (-4,4)
E_1_x_1 = k * (-q) * (X1-a) / ((X1-a)**2 + Z1**2)**(3/2) 
E_2_x_1 = k * (-q) * (X1+a) / ((X1+a)**2 + Z1**2)**(3/2) 
E_3_x_1 = k * (3*q) * X1 / (X1**2 + (Z1-b)**2)**(3/2) 
E_4_x_1 = k * (-q) * X1 / (X1**2 + (Z1+b)**2)**(3/2) 

E_total_x_1 = E_1_x_1 + E_2_x_1 + E_3_x_1 + E_4_x_1

E_1_z_1 = k * (-q) * Z1 / ((X1-a)**2 + Z1**2)**(3/2) 
E_2_z_1 = k * (-q) * Z1 / ((X1+a)**2 + Z1**2)**(3/2) 
E_3_z_1 = k * (3*q) * (Z1-b) / (X1**2 + (Z1-b)**2)**(3/2) 
E_4_z_1 = k * (-q) * (Z1+b) / (X1**2 + (Z1+b)**2)**(3/2) 

E_total_z_1 = E_1_z_1 + E_2_z_1 + E_3_z_1 + E_4_z_1

# Electric field contributions (-10,10)
E_1_x_2 = k * (-q) * (X2-a) / ((X2-a)**2 + Z2**2)**(3/2) 
E_2_x_2 = k * (-q) * (X2+a) / ((X2+a)**2 + Z2**2)**(3/2) 
E_3_x_2 = k * (3*q) * X2 / (X2**2 + (Z2-b)**2)**(3/2) 
E_4_x_2 = k * (-q) * X2 / (X2**2 + (Z2+b)**2)**(3/2) 

E_total_x_2 = E_1_x_2 + E_2_x_2 + E_3_x_2 + E_4_x_2

E_1_z_2 = k * (-q) * Z2 / ((X2-a)**2 + Z2**2)**(3/2) 
E_2_z_2 = k * (-q) * Z2 / ((X2+a)**2 + Z2**2)**(3/2) 
E_3_z_2 = k * (3*q) * (Z2-b) / (X2**2 + (Z2-b)**2)**(3/2) 
E_4_z_2 = k * (-q) * (Z2+b) / (X2**2 + (Z2+b)**2)**(3/2) 

E_total_z_2 = E_1_z_2 + E_2_z_2 + E_3_z_2 + E_4_z_2

# Streamplot for vector field (-4,4)
fig1 = plt.figure(figsize = (8,8))

plt.streamplot(x1, z1, E_total_x_1, E_total_z_1, linewidth=1, density=2, arrowstyle='->', arrowsize=1.5);
plt.title('Quesion 3: Numeric Electric Field Lines From -4 to 4')
plt.xlabel('x position (m)')
plt.ylabel('z position (m)')

# Streamplot for vector field (-10,10)
fig2 = plt.figure(figsize = (8,8))

plt.streamplot(x2, z2, E_total_x_2, E_total_z_2, linewidth=1, density=2, arrowstyle='->', arrowsize=1.5);
plt.title('Quesion 3: Numeric Electric Field Lines From -10 to 10')
plt.xlabel('x position (m)')
plt.ylabel('z position (m)')

# Streamplot simple vector field
E_x = 4 * k * q * b * Z2 * 3 * X2 / (X2**2+Z2**2)**2.5
E_z = 4 * k * q * b * (2*Z2**2 - X2**2) / (X2**2+Z2**2)**2.5
fig3 = plt.figure(figsize = (8,8))

plt.streamplot(x2, z2, E_x, E_z, linewidth=1, density=2, arrowstyle='->', arrowsize=1.5);
plt.title('Quesion 3: Approximate Electric Field Lines From -10 to 10')
plt.xlabel('x position (m)')
plt.ylabel('z position (m)')

fig1.savefig('Numeric_E_Field_1.png')
fig2.savefig('Numeric_E_Field_2.png')
fig3.savefig('Approx_E_Field_Far_Away.png')
```
